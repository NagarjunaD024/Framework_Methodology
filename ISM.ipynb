```python
import matplotlib.pyplot as plt
import numpy as np

# Example data: Driving Power and Dependence Power for factors based on dimensions
data = {
    'Factor': ['T1', 'T2', 'T3', 'T4', 'E1', 'E2', 'E3', 'O1', 'O2', 'O3'],
    'Driving Power': [18, 16, 15, 5, 14, 13, 7, 12, 6, 8],
    'Dependence Power': [5, 6, 14, 3, 12, 15, 16, 10, 14, 13]
}

# Cluster thresholds (average values)
avg_driving_power = 10
avg_dependence_power = 10

# Plotting the MICMAC analysis chart
fig, ax = plt.subplots(figsize=(10, 8))

# Plot clusters and data points
for i in range(len(data['Factor'])):
    x = data['Dependence Power'][i]
    y = data['Driving Power'][i]
    ax.scatter(x, y, label=data['Factor'][i])
    ax.text(x + 0.3, y, data['Factor'][i], fontsize=10)

# Draw average lines for division into clusters
ax.axhline(y=avg_driving_power, color='black', linestyle='--', linewidth=1)
ax.axvline(x=avg_dependence_power, color='black', linestyle='--', linewidth=1)

# Add quadrant labels
ax.text(0.5, avg_driving_power + 1, 'I Autonomous Cluster', fontsize=10, color='black')
ax.text(avg_dependence_power + 1, avg_driving_power + 1, 'II Dependent Cluster', fontsize=10, color='black')
ax.text(avg_dependence_power + 1, 0.5, 'III Linkage Cluster', fontsize=10, color='black')
ax.text(0.5, 0.5, 'IV Independent Cluster', fontsize=10, color='black')

# Customize plot
ax.set_title('MICMAC Analysis of Dimensions', fontsize=16)
ax.set_xlabel('Dependence Power', fontsize=12)
ax.set_ylabel('Driving Power', fontsize=12)
ax.set_xlim(0, 20)
ax.set_ylim(0, 20)
ax.legend(title='Factors', loc='upper left')
ax.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Save plot as an image
output_file = 'C:\\Users\\Nagarjuna\\Downloads\\micmac_analysis.png'
plt.tight_layout()
plt.savefig(output_file, dpi=300)
plt.show()

print(f"Graph saved as {output_file}")

```
